```python
 def sentimentalAnalysis(data,columnname):
        
        #downloading vader_lexicon for the process
        import matplotlib.pyplot as plt
        import seaborn as sns
        "Importing Necessary Packeage"
        from nltk.sentiment.vader import SentimentIntensityAnalyzer
        sid = SentimentIntensityAnalyzer()
        
        "Deleting null pr empty value"
        data.dropna(inplace=True)
        
        "Checking for a comment"
        #sid.polarity_scores(data[columnname][93])
        
        "Creating respective columns"
        
        data['scores'] = data[columnname].apply(lambda commentText: sid.polarity_scores(commentText))
        data['compound']  = data['scores'].apply(lambda score_dict: score_dict['compound'])
        data['Negtive']  = data['scores'].apply(lambda score_dict: score_dict['neg'])
        data['Postive']  = data['scores'].apply(lambda score_dict: score_dict['pos'])
        data['Neutral']  = data['scores'].apply(lambda score_dict: score_dict['neu'])
        
        "Creating final pos or neg using compound score"
        data['comp_score'] = data['compound'].apply(lambda c: 'pos' if c >=0 else 'neg')
        plt.clf()
        #comp=sns.countplot(x = 'comp_score', hue = 'Name', data = data, palette = 'magma')
        #comp.figure.savefig("date_charts.png")
        "Checking how many pos and neg"
        posneg=pd.DataFrame(data['comp_score'].value_counts())
        return posneg,data
    
```

```python
pos,data_Senti=sentimentalAnalysis(dataset,columnname='Chat')
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf=TfidfVectorizer(max_df=0.95,min_df=2,stop_words='english')
dtm=tfidf.fit_transform(df["Chat"])
feature_names = tfidf.get_feature_names_out()
```

```python
from sklearn.decomposition import LatentDirichletAllocation
# Number of topics
num_topics = 5

lda_model = LatentDirichletAllocation(n_components=num_topics, random_state=42)
lda_model.fit(dtm)
```

```python
for index, topic in enumerate(lda_model.components_):
    top_words = [feature_names[i] for i in topic.argsort()[-10:]]  # top 10 words
    print(f"Topic {index}: {', '.join(top_words)}")
```

```python
topic_results=lda_model.transform(dtm)
df["Topic"]=topic_results.argmax(axis=1)
```

```python
df['Date'] = pd.to_datetime(df['Date'], dayfirst=True,errors='coerce')

# --- normalize chat text ---
df['Chat_clean'] = df['Chat'].astype(str).str.strip().str.lower()
```

```python
#Summary Stats

# Who sent most messages
most_talkative = df['Name'].value_counts().idxmax()

# Most active date
most_active_date = df['Date'].value_counts().idxmax().date()

# Most active day of week
most_active_day = df['Date'].dt.day_name().value_counts().idxmax()

# Most active hour
df['Time'] = df['Time'].str.strip()  # remove leading/trailing spaces
df['Hour'] = pd.to_datetime(df['Time'], format='%I:%M %p').dt.hour
most_active_hour = df['Hour'].value_counts().idxmax()

# Avg messages per day
avg_messages = df.groupby('Date').size().mean()

#Deleted messages, Media shared
deleted_msgs = df[df['Chat'].str.contains("deleted", case=False)].shape[0]
media_msgs = df[df['Chat'].str.contains(".jpg|.mp4|.vcf", case=False)].shape[0]
```

```python
print("TIME ENCOUNTER")
print("Most Talkative Person:", most_talkative)
print("Most Active Date:", most_active_date)
print("Most Active Day of Week:", most_active_day)
print("Most Active Hour:", most_active_hour)
print("Average Messages per Day:", round(avg_messages, 2))

print("MEDIA ENCOUNTER")
print("Total Deleted Messages:", deleted_msgs)
print("Total Media Messages (images/videos/contacts):", media_msgs)
```

```python
# --- define regex patterns (order matters: check video -> voice -> generic) ---
import re
video_pat = re.compile(r'\bmiss(?:ed)?\b.*\bvideo\b.*\bcall\b')
voice_pat = re.compile(r'\bmiss(?:ed)?\b.*\bvoice\b.*\bcall\b')
generic_pat = re.compile(r'\bmiss(?:ed)?\b.*\bcall\b')  # fallback

def get_missed_type(text):
    if not isinstance(text, str):
        return None
    if video_pat.search(text):
        return 'video'
    if voice_pat.search(text):
        return 'voice'
    if generic_pat.search(text):
        return 'call'
    return None

df['missed_type'] = df['Chat_clean'].apply(get_missed_type)
df_missed = df[df['missed_type'].notna()].copy()

# --- try to infer who/what the missed event refers to (best-effort) ---
def infer_missed_by(row):
    txt = str(row['Chat_clean'])  # make sure it's a string
    
    # explicit phrasing where the owner is told "you missed"
    if 'you missed' in txt or 'you missed a' in txt:
        return 'You'
    
    # "missed voice call from <name>"
    m = re.search(r'missed.*call from\s+([^\:\,\.\n]+)', txt)
    if m:
        return m.group(1).strip()
    
    # "missed call by <name>" or "missed call - <name>"
    m2 = re.search(r'missed.*call (?:by|-\s*)([^\:\,\.\n]+)', txt)
    if m2:
        return m2.group(1).strip()
    
    # fallback ‚Üí Name column (sender)
    return row['Name']

# Apply row-wise and assign to single column
df_missed['missed_by'] = df_missed.apply(infer_missed_by, axis=1)


# --- quick summaries ---
print("Total missed events:", len(df_missed))
print("\nBy type:")
print(df_missed['missed_type'].value_counts())

print("\nTop 'missed_by' (who is shown as related to the missed event):")
print(df_missed['missed_by'].value_counts().head(10))

# Show the first few detected rows for manual verification
print("\nSample detected missed events:")
display(df_missed[['Date','Time','Name','Chat','missed_type','missed_by']].head(10))

# --- 5pivot counts per date and type (for plotting) ---
pivot = df_missed.pivot_table(index='Date', columns='missed_type', values='Chat', aggfunc='count', fill_value=0).sort_index()
print("\nPivot (missed calls per date/type):")
print(pivot)

# ---  plot stacked bar per date ---
if not pivot.empty:
    pivot.plot(kind='bar', stacked=True, figsize=(10,5))
    plt.title('Missed calls per day by type')
    plt.ylabel('Count')
    plt.xlabel('Date')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
else:
    print("No missed-call events found with the default patterns. Try expanding the regex (see notes).")
```
